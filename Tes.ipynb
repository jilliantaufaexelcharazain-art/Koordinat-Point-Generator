```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tkinter as tk
from tkinter import filedialog, messagebox

class PointGeneratorApp:
    def __init__(self, master):
        self.master = master
        master.title("Koordinat Point Generator")

        tk.Label(master, text="Jumlah titik:").grid(row=0, column=0)
        tk.Label(master, text="X_min:").grid(row=1, column=0)
        tk.Label(master, text="X_max:").grid(row=2, column=0)
        tk.Label(master, text="Y_min:").grid(row=3, column=0)
        tk.Label(master, text="Y_max:").grid(row=4, column=0)

        self.n_entry = tk.Entry(master)
        self.xmin_entry = tk.Entry(master)
        self.xmax_entry = tk.Entry(master)
        self.ymin_entry = tk.Entry(master)
        self.ymax_entry = tk.Entry(master)

        self.n_entry.grid(row=0, column=1)
        self.xmin_entry.grid(row=1, column=1)
        self.xmax_entry.grid(row=2, column=1)
        self.ymin_entry.grid(row=3, column=1)
        self.ymax_entry.grid(row=4, column=1)

        tk.Button(master, text="Generate", command=self.generate_points).grid(row=5, column=0, columnspan=2, pady=10)

    def generate_points(self):
        try:
            n = int(self.n_entry.get())
            xmin = float(self.xmin_entry.get())
            xmax = float(self.xmax_entry.get())
            ymin = float(self.ymin_entry.get())
            ymax = float(self.ymax_entry.get())

            x = np.random.uniform(xmin, xmax, n)
            y = np.random.uniform(ymin, ymax, n)

            data = pd.DataFrame({'X': x, 'Y': y})

            # Simpan ke file TXT
            filename = filedialog.asksaveasfilename(defaultextension=".txt",
                                                    filetypes=[("Text Files", "*.txt")])
            if filename:
                data.to_csv(filename, index=False, sep='\t')
                messagebox.showinfo("Sukses", f"Data berhasil disimpan ke {filename}")

            # Plot hasil
            plt.scatter(x, y, c='blue')
            plt.title("Scatter Plot Koordinat")
            plt.xlabel("X")
            plt.ylabel("Y")
            plt.grid(True)
            plt.show()

        except Exception as e:
            messagebox.showerror("Error", str(e))

root = tk.Tk()
app = PointGeneratorApp(root)
root.mainloop()
```
